```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

```python
total_players=len(purchase_data.SN.unique()) #to count unique value and count the length #2 ways to prefer to column: this way when there is no space 
total_players_df= pd.DataFrame({"Total Players": [total_players]})
total_players_df
```

## Purchasing Analysis (Total)

```python
#number of unique items
Unique_items = len(purchase_data['Item Name'].unique()) #way to refer to column when there is space
#average price
Average_price = purchase_data['Price'].mean()

#Total number of purchases
No_of_Purchases = len(purchase_data["Purchase ID"].value_counts())
#Total Purchase Price
Total= purchase_data.Price.sum()

#Summary data frame
Purchasing_Analysis= pd.DataFrame({'Number of Unique Items': [Unique_items],  #Use square bracket because it is list.
                                 'Average Price': [Average_price],            #ko phai 1 series cua data ma chi co 1 ket qua. list vs dataframe   
                                   'Number of Purchases': [No_of_Purchases],
                                  'Total Revenue':[Total]})
#format for currency
Purchasing_Analysis["Average Price"] = Purchasing_Analysis["Average Price"].map("${:,.2f}".format)
Purchasing_Analysis["Total Revenue"] = Purchasing_Analysis["Total Revenue"].map("${:,.2f}".format)

#Another way to format
    #summary_df.style.format({'Average Price':"${:,.2f}",
                             #'Total Revenue': '${:,.2f}'})
Purchasing_Analysis
```

## Gender Demographics

```python
#drop duplicate screen name
purchase_data_dropdup= purchase_data.drop_duplicates(['SN'])
    #purchase_data_dropdup.info()
#no of players by gender
players_demo = purchase_data_dropdup.Gender.value_counts()
players_demo

#total no of players
total_players=len(purchase_data.SN.unique()) #from above #1
#demo percentages
players_percent = players_demo / total_players *100
#summary table
gender_demo = pd.DataFrame({"Total Count": players_demo,"Percentage of Players": players_percent}) 
                                                                #vi la dataframe nen ko can square bracket
#format for percentage
gender_demo["Percentage of Players"] = gender_demo["Percentage of Players"].map("{:,.2f}%".format)
gender_demo


```


## Purchasing Analysis (Gender)

```python
#group by gender column
gender_group_by= purchase_data.groupby(["Gender"]) #do not use drop duplicate here because it will mess with the total

#use a data function to visualize "Groupby"
    #gender_group_by.count().head(10)

#Purchase Count per gender
purchase_count_demo = gender_group_by["Purchase ID"].count()

#Average Purchase Price per gender
Average_price_per_gender =  gender_group_by["Price"].mean()

#Total purchase value per gender
Total_price_per_gender =  gender_group_by["Price"].sum()

#Avg Total Purchase per Person
Avg_per_person_per_gender = Total_price_per_gender / players_demo #players_demo from above where we already dropped duplicate value to count no of male/female

#summary chart
purchasing_analysis_gender = pd.DataFrame({'Purchase Count' :purchase_count_demo,
                                           'Average Purchase Price':Average_price_per_gender,
                                          'Total Purchase Value':Total_price_per_gender,
                                          'Avg Total Purchase per Person':Avg_per_person_per_gender})
#format currency
purchasing_analysis_gender["Average Purchase Price"] = purchasing_analysis_gender["Average Purchase Price"].map("${:,.2f}".format)
purchasing_analysis_gender["Total Purchase Value"] = purchasing_analysis_gender["Total Purchase Value"].map("${:,.2f}".format)
purchasing_analysis_gender["Avg Total Purchase per Person"] = purchasing_analysis_gender["Avg Total Purchase per Person"].map("${:,.2f}".format)
purchasing_analysis_gender



```

## Age Demographics

```python
bins= [0,9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 39.9,999]
# Create labels for these bins
group_labels = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34","35-39", "40+"]
# Slice the data and place it into bins
pd.cut(purchase_data["Age"], bins, labels=group_labels).head()

#add new column to DataFrame
purchase_data_age_sort=purchase_data #to define purchase_data_age_sort and dont mess with original data
purchase_data_age_sort['Age Ranges'] = pd.cut(purchase_data["Age"], bins, labels=group_labels)

#Groupby age ranges
age_group_by = purchase_data_age_sort.groupby("Age Ranges")
#use a data function to visualize "Groupby"
    #age_group_by.count().head(10)
#count total of unique SN per age ranges
unique_SN_per_age = age_group_by['SN'].nunique()

#percentage of players per age ranges

percentage_unique_age = unique_SN_per_age / total_players *100

#summary table

players_count_analysis = pd.DataFrame({'Total Count':unique_SN_per_age,'Percentage of Players':percentage_unique_age})

#format percentage
players_count_analysis['Percentage of Players'] =players_count_analysis['Percentage of Players'].map("{:,.2f}%".format)

# Format the data frame with no index name in the corner
players_count_analysis.index.name = None #remove Age Ranges header. try to remove this row and see
players_count_analysis

```

## Purchasing Analysis (Age)

```python
#Purchase Count per age range
purchase_count_age = age_group_by["Purchase ID"].count()

#Average Purchase Price per age range
Average_price_per_age = age_group_by["Price"].mean()

#Total purchase value per age range
Total_price_per_age =  age_group_by["Price"].sum()

#Avg Total Purchase per Person
Avg_per_person_per_age = Total_price_per_age / unique_SN_per_age #from above

#summary chart 
purchasing_analysis_age = pd.DataFrame({'Purchase Count' :purchase_count_age, #we dont use square bracket because of dataframe
                                           'Average Purchase Price':Average_price_per_age,
                                          'Total Purchase Value':Total_price_per_age,
                                          'Avg Total Purchase per Person':Avg_per_person_per_age})
#format currency
purchasing_analysis_age["Average Purchase Price"] = purchasing_analysis_age["Average Purchase Price"].map("${:,.2f}".format)
purchasing_analysis_age["Total Purchase Value"] = purchasing_analysis_age["Total Purchase Value"].map("${:,.2f}".format)
purchasing_analysis_age["Avg Total Purchase per Person"] = purchasing_analysis_age["Avg Total Purchase per Person"].map("${:,.2f}".format)
purchasing_analysis_age


```

## Top Spenders

```python
#group by SN column
SN_group_by =  purchase_data.groupby('SN')
#use a data function to visualize "Groupby"
    #SN_group_by.count().head(10)

#Number of purchase per SN
SN_purchase_count = SN_group_by['Purchase ID'].count()

#Average purchase price
SN_average_price = SN_group_by['Price'].mean()

#Total Purchase Value
SN_total_price = SN_group_by['Price'].sum()

#summary table
SN_analysis = pd.DataFrame({'Purchase Count': SN_purchase_count, 'Average Purchase Price': SN_average_price,
                           'Total Purchase Value': SN_total_price})

#sort per total purchase value
SN_analysis = SN_analysis.sort_values('Total Purchase Value', ascending =False)

#format currency
SN_analysis["Average Purchase Price"]= SN_analysis['Average Purchase Price'].map("${:,.2f}".format)
SN_analysis["Total Purchase Value"]=SN_analysis['Total Purchase Value'].map("${:,.2f}".format)
#remove SN index# Format the data frame with no index name in the corner
SN_analysis.index.name = None #remove SN header. try to remove this row and see
SN_analysis.head()
```

## Most Popular Items

```python
purchase_data.head()
#Retrieve Item ID, Item Name, and Item Price 
most_popular_items= purchase_data[["Item ID","Item Name","Price"]]
#most_popular_items.head()
#Rename columns
#most_popular_items=most_popular_items.rename(columns={"Purchase ID":"Item ID","Price":"Item Price"})
#Grioup by Item ID and Item Name
sort_most_pop_items =most_popular_items.groupby(['Item ID','Item Name'])
#use a data function to visualize "Groupby"
#sort_most_pop_items.count().head(10)

#Purchase count per item

purchase_count_per_item = sort_most_pop_items['Item ID'].count()

#Total purchase value per item
Total_purchase_per_item = (sort_most_pop_items['Price'].sum())

#Item price
Price_per_item = Total_purchase_per_item / purchase_count_per_item

#summary table
most_pop_items_analysis = pd.DataFrame ({'Purchase Count':purchase_count_per_item, "Item Price" :Price_per_item,
                                        'Total Purchase Value': Total_purchase_per_item })


#Sort the purchase count column in descending order
most_pop_items_analysis_sort=most_pop_items_analysis.sort_values('Purchase Count',ascending=False)
#format currency
most_pop_items_analysis_sort['Item Price']=most_pop_items_analysis['Item Price'].map("${:,.2f}".format)
most_pop_items_analysis_sort['Total Purchase Value']=most_pop_items_analysis['Total Purchase Value'].map("${:,.2f}".format)
most_pop_items_analysis_sort.head()


```
